## Exploratory Data Analysis

**Exploratory Data Analysis (EDA)** is the process of investigating and visualizing raw datasets to uncover their main characteristics, patterns, anomalies, and relationships. It's a crucial first step in any data science project before modeling.

### Key Goals of EDA:
- **Understand the data structure**: Shape, data types, missing values
- **Identify patterns & trends**: Distributions, correlations, outliers
- **Generate insights**: Find relationships between variables
- **Detect data quality issues**: Missing values, duplicates, inconsistencies
- **Inform modeling**: Choose appropriate algorithms based on data characteristics

### Common EDA Techniques:
- **Descriptive statistics**: Mean, median, std, min, max
- **Data visualization**: Histograms, boxplots, scatter plots, heatmaps
- **Correlation analysis**: Identify relationships between features
- **Outlier detection**: Identify and understand anomalies
- **Missing data assessment**: Evaluate gaps and imputation needs

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define features and targets from the regression problem
features = ['PT08.S1(CO)', 'PT08.S2(NMHC)', 'PT08.S3(NOx)', 'PT08.S4(NO2)', 
            'PT08.S5(O3)', 'T', 'RH', 'AH']
targets = ['CO(GT)', 'NMHC(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)']

# Combine features and targets
all_columns = features + targets

# Select only these columns that exist in the dataframe
selected_columns = [col for col in all_columns if col in df.columns]

# Calculate correlation matrix
correlation_matrix = df[selected_columns].corr()

# Create visualization 1: Full correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Full Correlation Analysis (All Features and Targets)')
plt.tight_layout()
plt.show()
```

```python
## Outliers and Distribution Analysis

# Display distributions and detect outliers for features
print("FEATURES - Distribution and Outlier Analysis:")
print("="*70)

fig, axes = plt.subplots(len(features), 1, figsize=(14, 4*len(features)))

for idx, feature in enumerate(features):
    if feature in df.columns:
        data = df[feature].dropna()
        
        # Histogram
        axes[idx].hist(data, bins=30, edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'{feature} - Distribution')
        axes[idx].set_ylabel('Frequency')
        
        # Calculate outliers using IQR method
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data < lower_bound) | (data > upper_bound)]
        
        print(f"\n{feature}:")
        print(f"  Mean: {data.mean():.2f}, Std: {data.std():.2f}")
        print(f"  Min: {data.min():.2f}, Max: {data.max():.2f}")
        print(f"  Outliers count: {len(outliers)} ({len(outliers)/len(data)*100:.2f}%)")

plt.tight_layout()
plt.show()

# Display distributions and detect outliers for targets
print("\n\nTARGETS - Distribution and Outlier Analysis:")
print("="*70)

fig, axes = plt.subplots(len(targets), 1, figsize=(14, 4*len(targets)))

for idx, target in enumerate(targets):
    if target in df.columns:
        data = df[target].dropna()
        
        # Histogram
        axes[idx].hist(data, bins=30, edgecolor='black', alpha=0.7, color='orange')
        axes[idx].set_title(f'{target} - Distribution')
        axes[idx].set_ylabel('Frequency')
        
        # Calculate outliers using IQR method
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data < lower_bound) | (data > upper_bound)]
        
        print(f"\n{target}:")
        print(f"  Mean: {data.mean():.2f}, Std: {data.std():.2f}")
        print(f"  Min: {data.min():.2f}, Max: {data.max():.2f}")
        print(f"  Outliers count: {len(outliers)} ({len(outliers)/len(data)*100:.2f}%)")

plt.tight_layout()
plt.show()
```

## Handling Missing Data

In this dataset, missing values are represented as **-200**. There are several strategies to handle missing data:

### 1. **Deletion (Removal)**
- **Drop entire rows**: Remove rows containing missing values
- **Drop entire columns**: Remove columns with too many missing values
- **Pros**: Simple, no assumptions made
- **Cons**: Loss of data, may bias results if data is not missing randomly

### 2. **Imputation (Filling)**
- **Mean/Median/Mode imputation**: Replace missing values with the mean (continuous), median, or mode (categorical)
  - Mean: Good for normally distributed data
  - Median: Robust to outliers
  - Mode: For categorical data
- **Forward/Backward fill**: For time-series data, fill with previous or next value
- **Pros**: Retains sample size
- **Cons**: Reduces variance, creates artificial patterns

### 3. **Advanced Imputation Methods**
- **KNN Imputation**: Use K-Nearest Neighbors to fill values based on similar records
- **Regression Imputation**: Predict missing values using other features
- **Multiple Imputation**: Create multiple datasets with different imputed values for uncertainty analysis
- **Pros**: More sophisticated, retains relationships
- **Cons**: Computationally expensive, requires careful tuning

### 4. **Create Missing Indicator**
- Add a binary column indicating whether data was missing
- Replace missing values with mean/median
- **Pros**: Preserves information about missingness
- **Cons**: Adds dimensionality

### 5. **Model-Based Handling**
- Some algorithms (e.g., XGBoost) can handle missing values internally
- **Pros**: Leverages algorithm capabilities
- **Cons**: Limited to specific models

### Recommended Approach for This Dataset:
1. **Replace -200 with NaN** to properly identify missing values in pandas
2. **Analyze missingness pattern** (is it random or systematic?)
3. **Choose strategy**:
   - If <5% missing: Consider deletion
   - If 5-30% missing: Use imputation (mean/median or KNN)
   - If >30% missing: Consider removing that feature or using advanced methods

```python
# Replace -200 with NaN
df_cleaned = df.copy()
df_cleaned.replace(-200, pd.NA, inplace=True)

# Select only features and targets
all_features_targets = features + targets
df_cleaned = df_cleaned[[col for col in all_features_targets if col in df_cleaned.columns]]

print("="*70)
print("MISSINGNESS ANALYSIS - FEATURES AND TARGETS ONLY")
print("="*70)

# Missingness pattern by column
print("\nMissingness by Column:")
print("-"*70)
missing_by_column = df_cleaned.isnull().sum().sort_values(ascending=False)
missing_percent_by_column = (df_cleaned.isnull().sum() / len(df_cleaned)) * 100
missing_summary = pd.DataFrame({
    'Column': missing_by_column.index,
    'Missing_Count': missing_by_column.values,
    'Missing_Percent': missing_percent_by_column[missing_by_column.index].values
})
print(missing_summary)

# Proportion of rows with at least one missing value
print("\n" + "="*70)
print("ROWS WITH MISSING VALUES:")
print("-"*70)
rows_with_missing = df_cleaned.isnull().any(axis=1).sum()
total_rows = len(df_cleaned)
proportion_rows_missing = (rows_with_missing / total_rows) * 100

print(f"Total rows: {total_rows}")
print(f"Rows with at least one missing value: {rows_with_missing}")
print(f"Proportion of rows with missing values: {proportion_rows_missing:.2f}%")
print(f"Rows with complete data: {total_rows - rows_with_missing} ({100-proportion_rows_missing:.2f}%)")

```
